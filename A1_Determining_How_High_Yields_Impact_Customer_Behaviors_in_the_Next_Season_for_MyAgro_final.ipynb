Our client myAgro is committed to empowering smallholder farmers, especially women, to reduce poverty and increase food security. They have served over 115,000 farmers and 52% of farmers are women.

MyAgro uses a mobile layaway savings model that enables smallholder farmers in West Africa to invest their own funds in high-quality seed, fertilizer, and agricultural training to increase their harvests and income. With a business model that utilizes a mobile layaway plan—allowing farmers to save earnings from harvest time to reinvest in the planting season.

MyAgro's business cycle starts with a sales season with enrolling clients and ordering packages, followed by the delivery season, and payments are collected over the farming season.

## <b> 5.3 Data cleaning and data manipulation </b>

Client_code is the unique key that defines each client and is used to join and filter across the tables. Because the client code column in the harvest table is datatype int, and all the other tables' client code columns are in objects, we convert them all to string to perform accurate merging and filtering.

```python
# convert client_code to str to match client_code column in all files
harvest["client_code"] = harvest["client_code"].astype(str)
client["code"] = client["code"].astype(str)
package["client_code"] = package["client_code"].astype(str)
payment["client_code"] = payment["client_code"].astype(str)
```

#### <b> Client table</b>

In client table, we manipulated data is the following steps:


1.   We renamed the "code" column to "client_code" so it matches other tables.
2.   We removed clients that have not enrolled in any season.

In total, we ended up with our client_cleaned table of 317030 clients.


```python
# rename "code" col to "client_code" to match the key in all datasets
client.rename(columns = {'code': 'client_code'}, inplace = True)
# remove the clients who never enrolled
client_cleaned = client.dropna(subset=['most_recent_year_enrolled'], how='all')
```

```python
client_cleaned.info()
```

#### <b> Harvest table  </b>

We decide to fill null values in columns "planted_using_myagro_semoir", "staff_measured_plot", and "staff_present_fertilizer" with 0 because they are boolean columns with 1 indicating True, and we assume that no record indicates False.

We did not fill null in "trainings_attened" with 0 because there are a lot of null values in 2020 and 2022, and we cannot decide if it is an incomplete data collection. 

```python
#Replacing the nulls
harvest.fillna({"planted_using_myagro_semoir":0,
                          "staff_measured_plot":0,
                          "staff_present_fertilizer":0}, 
                         inplace=True)
harvest.info()
```

The client table contains all documented clients within the past 10 years of MyAgro's business. As we are focusing on yields and our harvest data only contains 3181 data entries and are from 2020 to 2022, we will use an inner join to filter out our desired clients.

```python
client_with_harvest=client_cleaned.merge(harvest,how='inner',on=["client_code"])
client_with_harvest.head(3)
```

1.   "year_received" indicated which year the payment was received.

### <b> 6.1.1 Seniority of all clients on record </b>

By looking at the seniority of clients registered from 2021 to 2023 in client table, we answer the following questions:

##### <b> Q1. How do seniority of clients change over the years? </b>



**Observation**: To understand the results from 2023, we need to take into consideration that the sales season for 2023 is ongoing, so it is very likely that there will be more clients enrolled for season 2023 in the future.<br>
 From the comparison bar chart, we can see that the total number of clients enrolled each year is increasing over the past 3 years. The majority of clients enrolled each season are returning clients. For season 2022, MyAgro was successful to attract many new clients, and this trend seems to continue with season 2023 based on our observation of the ratio between new and returning clients. We will need more information to reach that conclusion.



```python
yr_lst = pd.unique(client_cleaned["most_recent_year_enrolled"])
count_new = []
count_return = []
for yr in yr_lst: # only 3 years so 3 iters which does not affect code performance
  #count rows where "most_recent_year_enrolled" is the yr selected and seniority is new
  count_new.append(len(client_cleaned[(client_cleaned['most_recent_year_enrolled']==yr) & (client_cleaned['seniority']=='new')]))
  #count rows where "most_recent_year_enrolled" is the yr selected and seniority is returning
  count_return.append(len(client_cleaned[(client_cleaned['most_recent_year_enrolled']==yr) & (client_cleaned['seniority']=='returning')]))

d = {"most recent year enrolled":yr_lst, "new clients":count_new,"returning clients":count_return}
seniority_over_yr = pd.DataFrame(data=d)
seniority_over_yr.sort_values(by=["most recent year enrolled"],inplace=True,ignore_index=True)
print(seniority_over_yr)
```

```python
sns.set_palette('crest')
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
# comparison bar chart
N = 3
ind = np.arange(N) 
width = 0.25
fig, ax = plt.subplots()

new = seniority_over_yr["new clients"]
returning = seniority_over_yr["returning clients"]

bar1 = ax.bar(ind-width/2, new, width, label='new clients')
bar2 = ax.bar(ind+width/2, returning, width, label='returning clients')

ax.set_xlabel("most recent year enrolled")
ax.set_ylabel('client counts')
ax.set_title("Seniority over years")  
ax.set_xticks(ind)
ax.set_xticklabels(['2021','2022','2023'])

autolabel(bar1)
autolabel(bar2)

ax.legend()

fig.tight_layout()
plt.show()
```

##  **6.2 Benchmarking clients with high yields and analyzing potential factors that lead to high yields**

### <b> 6.2.1. Define "high yields" </b> 

We define "high yields" as yields that are higher than the average yield "for each crop in each country each year" because yields are different based on crops, country, and seasons (which represent the weather and precipitation). 

**Calculate the average yield per crop per country per year**

```python
client_with_harvest['yield_mean']=client_with_harvest.groupby(["season","crop","country_name"])["yield"].transform('mean')
client_with_harvest.head(3)
```

**Find customer with high yields**

We are adding a boolean column to "high yield." And we are marking clients with high yield with 1, and else with 0. 

```python
client_with_harvest['high_yield']=" "

def function(a,b):
  if a>b: return 1
  else: return 0

client_with_harvest['high_yield']=client_with_harvest.apply(lambda x:function(x['yield'],x['yield_mean']),axis=1)
client_with_harvest.head(3)
```

## **6.3 Analyzing how customer behaviors change in the next season for clients with high yields**

In this section, we look at how customer behaviors change in the next season after they have high yields in the current season. We are considering 2 metrics, which are payments made for the next season, and packages ordered. For packages, we are considering both the number of packages and the price of packages.

We have harvest data for 2020, 2021, and 2022 seasons. However, because we don't have the full sales data for the 2023 season, which is ongoing for orders, we are not able to utilize 2022 harvest data. So we will consider 2020 and 2021 harvest data and the corresponding 2021, 2022 sales data.

### **Overall Summary of key observations**


*   High yields clients had a quit rate of 17% in 2020, and 60% in 2021, which indicates that from our data, high yields do not increase customer loyalty to the program.
*   High yields clients mostly chose to order what they had ordered or order less products in the following season. This indicates that high yields do not lead into more purchases.
*   Our observations are not logical nor reasonable, which indicates that there are underlying data issues, and more researches and contacts with farmers are needed to understand the causes of our results.



### **6.3.1 How does high yield affect clients' payments in the next season?**

#### **2020-2021**

In this section, we are analyzing how clients with high yields in 2020's behaviors change in 2021 sales season.

We start by filtering for high yields client data in 2020.

There are 308 clients with high yields in 2020.

```python
high_yield2020 = client_with_harvest.loc[(client_with_harvest["high_yield"]==1) & (client_with_harvest["season"]==2020)]["client_code"]
print("There are",len(high_yield2020.unique()),"clients with high yields in 2020.")
```

## **6.4 Challenges and potential explanations of results** 

Following are potential explanation we have from our research and MyAgro staffs about the discrepencies and irregular quit rates we observed from our analysis.

*   Yield data were collected with self-reported survey giving to farmers. From our analysis in 6.2, we can see that **self-reported yield are significantly higher than yields measured by MyAgro staffs**. This could lead to error in locating high yield clients.  
*   High quit rates in general, even for clients with high yields in the prior season: As per MyAgro's suggestion, they are facing a data quality issue. Their contracted VE (village entrepreneurs) who are in charge of enrolling clients into the program have **assigned returning clients with new client codes**, which made clients record discontinuous. And hense hard to filter clients with high yield harvest results in multiple years. It is very likely that in real life, there are more clients with high yields in multiple years.
*   Large decrease in enrollment and product purchased between 2021 and 2022. Due to the war between Russia and Ukraine, **fertilisers' prices increased drastically** in west Africa, as potassium fertilizers are mostly imported from Russia. Many clients decided to quit the program and purchased seeds from other companys because they can no longer afford it.
